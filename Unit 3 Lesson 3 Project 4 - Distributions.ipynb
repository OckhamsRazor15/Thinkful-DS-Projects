```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
r1 = np.random.normal(50, 300, 100)
r2 = np.random.poisson(1, 100)
r3 = np.random.binomial(1, .5, 100)
r4 = np.random.binomial(20, 0.5, 100)
r5 = np.random.gamma(5,1, 100)
r6 = np.random.wald(3, 2, 100)
```

```python
plt.hist(r1)
plt.axvline(r1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(r1.mean() + r1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(r1.mean() - r1.std(), color='b', linestyle='dashed', linewidth=2)
```
