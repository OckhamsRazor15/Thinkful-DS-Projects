```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
%matplotlib inline
```

```python
pl = pd.read_csv("../pl1718.csv")
```

```python
home_city = []
away_city = []
home_lfc = []
away_lfc = []
#FTR = Full Time Result, FTHG = Full Time Home Goals, FTAG = Full Time Away Goals
teams = pl.loc[:, ('Date','HomeTeam', 'AwayTeam','FTR', 'FTHG' ,'FTAG')]

#iterate over the teams df and pull out all rows related to either Man City or Liverpool. 
# splits them into home and away game sets
for index, row in teams.iterrows():
    if row['HomeTeam'] == 'Man City':
        home_city.append(row)
    elif row['HomeTeam'] == 'Liverpool':
        home_lfc.append(row)

    if row['AwayTeam'] == 'Man City':
        away_city.append(row)
    elif row['AwayTeam'] == 'Liverpool':
        away_lfc.append(row)
```

```python
#go through and pull out goals and put them into a list
home_goals_city = []
for hw in home_city:
    home_goals_city.append(hw[4])
    
home_goals_lfc = []
for hw in home_lfc:
    home_goals_lfc.append(hw[4])
    
away_goals_city = []
for hw in away_city:
    away_goals_city.append(hw[5])
#the data didn't have the last game, they scored 1
away_goals_city.append(1)

away_goals_lfc = []
for hw in away_lfc:
    away_goals_lfc.append(hw[5])
    
home_goals_against_city = []
for hw in home_city:
    home_goals_against_city.append(hw[5])

home_goals_against_lfc = []
for hw in home_lfc:
    home_goals_against_lfc.append(hw[5])
    
away_goals_against_city = []
for hw in away_city:
    away_goals_against_city.append(hw[4])
away_goals_against_city.append(0)

away_goals_against_lfc = []
for hw in away_lfc:
    away_goals_against_lfc.append(hw[4])
```

```python
# Comparing Liverpool to Man City
plt.figure(figsize=[20,5])

plt.subplot(1,2,1)
plt.plot(home_goals_city, color='blue', alpha=.5, label="Man City")
plt.plot(home_goals_lfc, color='red', alpha=.5, label="LFC")
plt.title("Home Goals Scored")
plt.xlabel("Game # Home")
plt.xlim(0,18)
plt.ylabel("Goals Scored")
plt.legend()


plt.subplot(1,2,2)
plt.plot(away_goals_city,color='blue', alpha=.5, label="Man City")
plt.plot(away_goals_lfc, color='red', alpha=.5, label="LFC")
plt.title("Away Goals Scored")
plt.xlabel("Game # Away")
plt.xlim(0,18)
plt.ylabel("Goals Scored")
plt.legend()

plt.figure(figsize=[20,5])

plt.subplot(1,2,1)
plt.plot(home_goals_against_city, color='blue', alpha=.5, label="Man City")
plt.plot(home_goals_against_lfc, color='red', alpha=.5, label="LFC")
plt.title("Home Goals Conceded")
plt.xlabel("Game # Home")
plt.xlim(0,18)
plt.ylabel("Goals Conceded")
plt.legend()


plt.subplot(1,2,2)
plt.plot(away_goals_against_city,color='blue', alpha=.5, label="Man City")
plt.plot(away_goals_against_lfc, color='red', alpha=.5, label="LFC")
plt.title("Away Goals Conceded")
plt.xlabel("Game # Away")
plt.xlim(0,18)
plt.ylabel("Goals Conceded")
plt.legend()

#looking at variances/avgs in goals
print("home goals")
print(np.var(home_goals_city))
print(np.var(home_goals_lfc))
print('---------------')
print(np.mean(home_goals_city))
print(np.mean(home_goals_lfc))
print('---------------')

print("home goals against")
print(np.var(home_goals_against_city))
print(np.var(home_goals_against_lfc))
print('---------------')
print(np.mean(home_goals_against_city))
print(np.mean(home_goals_against_lfc))
print('---------------')

print("away goals")
print(np.var(away_goals_city))
print(np.var(away_goals_lfc))
print('---------------')
print(np.mean(away_goals_city))
print(np.mean(away_goals_lfc))
print('---------------')

print("away goals against")
print(np.var(away_goals_against_city))
print(np.var(away_goals_against_lfc))
print('---------------')
print(np.mean(away_goals_against_city))
print(np.mean(away_goals_against_lfc))
```
